```python
def age_nan(cols):
    Age = cols[0]
    Pclass = cols[1]
    if pd.isnull(Age):
        if Pclass ==1:
            Age = 37
        elif Pclass == 2:
            Age = 29
        else:
            Age = 24
    return Age
    
train['Age']=train[['Age','Pclass']].apply(age_nan,axis = 1)
```

```python
sns.heatmap(train.isnull())
```

```python
train.drop('Cabin',axis = 1,inplace = True)
```

```python
train.head()
```

```python
train.info()
```

```python
#dummy variables
sex = pd.get_dummies(train['Sex'],drop_first = True)

#by using drop_first we get one less DUMMY VARIABLE for a categorical variable
```

```python
embark = pd.get_dummies(train['Embarked'],drop_first = True)
```

```python
train.head()
```

```python
train = pd.concat([train,sex,embark],axis = 1)
```

```python
train.drop(columns=['Name','Sex','Ticket','Embarked'],axis = 1,inplace = True)
```

```python
train.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(train.drop('Survived',axis=1), 
                                                    train['Survived'], test_size=0.20, 
                                                    random_state=101)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logmodel = LogisticRegression()
```

```python
logmodel.fit(X_train,y_train)
```

```python
predictions = logmodel.predict(X_test)
```

```python
predictions
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
cm  = confusion_matrix(y_test,predictions)
```

```python
cm
```

```python
print(classification_report(y_test,predictions))
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf_model = RandomForestClassifier(n_estimators = 25)
```

```python
rf_model.fit(X_train,y_train)
```

```python
prediction_rf = rf_model.predict(X_test)
```

```python
cm_rf = confusion_matrix(y_test,prediction_rf)
```

```python
cm_rf
```

```python
print(classification_report(y_test,prediction_rf))
```

```python
#so here greatest accuracy in random forest classifier . so it will be used for test
```

```python
#for finding the ideal number of estimators
```

```python
error_rate = []
for i in range(1,51):
    rf_model = RandomForestClassifier(n_estimators = i)
    rf_model.fit(X_train,y_train)
    pred_i = rf_model.predict(X_test)
    error_rate.append(np.mean(pred_i!=y_test))
```

```python
plt.figure(figsize = (10,6))
plt.plot(range(1,51),error_rate,linestyle= 'dashed',marker= 'o',markerfacecolor = 'red',markersize = '8')
plt.title('Error Rate vs N_estimator Value')
plt.xlabel('N_estimator value or K')
plt.ylabel('Error Rate')
```

```python
error_rate[37]
```

```python
error_rate[44]
```

```python
#so both are same and hence we can choose any value of K
rf = RandomForestClassifier(n_estimators= 37)
```

```python
rf.fit(X_train, y_train)
```

```python
prediction_rf = rf.predict(X_test)
```

```python
cm_rf = confusion_matrix(y_test,prediction_rf)
```

```python
cm_rf
```

```python
print(classification_report(y_test,prediction_rf))
```
